Uçuşa qalmış günlərə əsasən biletlərin satış qiymətlərini sıralayırıq.

```python
df.groupby('days_left')['price'].mean().sort_values(ascending=False).head(15)
```

Maksimum qiymətləri analiz edək. Görürük ki, uçuşa az qalmış günlərin satışında bahalaşma görünür, amma bu linear(xətti) deyil.

```python
df.groupby('days_left').agg({'price' : max}).head(10)
```

```python
df
```

Economy və business sinifə görə qiymət ortalaması arasında qat-qat fərqi görürük.

```python
df.groupby(['class'])['price'].mean()
```

```python
df.groupby(['airline','class'])['price'].mean().unstack
```

```python
mean_of_duration = df['duration'].mean()
print(mean_of_duration.round(2))
```

```python
df.groupby('class')['duration'].mean()
```

Durak saylarının ortalamalarının qiymətə necə təsir etdiyini görürük.

```python
df.groupby('stops')['price'].mean().unstack
```

```python
df.groupby('stops')['class'].value_counts().reset_index()
```

```python
df.groupby('stops')['duration'].agg('mean').sort_values(ascending=True).round(2)
```
